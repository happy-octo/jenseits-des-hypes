```python
from phi.agent import Agent
from phi.model.ollama import Ollama
```

```python
ollama_host="http://ollama-predictor.01-ollama.svc.cluster.local:11434"

extract_agent = Agent(
    model=Ollama(id="sroecker/nuextract-tiny-v1.5", host=ollama_host),
    description="You extract information.",
)

def predict_nuextract(input_text):
    template = """
    {
        "Customer": {
            "Name": "",
            "Address": "",
            "Policy Number": "",
            "Telephone Number": "",
        },
        "Case": {
            "Accident Location": ""
            "Time": ""
        }
    }
    """
    template = f"""<|input|>\n ### Template:\n{template}\n### Text:\n{input_text}\n\n<|output|>"""

    return template
```
